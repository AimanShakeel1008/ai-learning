## Part 1 — Why we use libraries at all\n\nA **library** is a chunk of code someone else already wrote, tested, and packaged up, so you can `import` it and use it instead of writing that same code yourself from scratch.\n\nHere's a tiny, honest demo. Python's built-in `math` library can compute a square root. But a square root is not magic — it is just a number-finding process, and we can write our own version of that process by hand in a few lines, using an old technique called **Newton's method**: start with a guess, then repeatedly average the guess with `x / guess`, and the guess gets closer to the true square root each time.\n\nThe next cell defines that hand-rolled version."

```python
def my_sqrt(x, guess=1.0, steps=10):\n    for _ in range(steps):\n        guess = (guess + x / guess) / 2\n    return guess\n\n\nprint(my_sqrt(2))
```

Now compare that to Python's built-in `math` library doing the exact same job:"

```python
import math\n\nprint(math.sqrt(2))
```

Both lines print the same number. So why bother with `math` at all, if a hand-rolled version works?\n\nBecause `my_sqrt` only *looks* finished. It breaks the moment you feed it something it wasn't built for — negative numbers, zero, or a truly enormous number — and nobody has stress-tested it against those cases. `math.sqrt` has been used by millions of programs for decades, so those edge cases were found and fixed long ago. A library is not just \"someone else's code\" — it is code that has already survived contact with far more real-world cases than anything one person writes in an afternoon.\n\nThis is exactly why AI work leans on a handful of specific libraries instead of hand-rolling everything, the same way `my_sqrt` was hand-rolled above:\n\n- **NumPy** — fast math on big grids of numbers (called **arrays**). Lesson 05 is entirely about this.\n- **pandas** — loading and working with data that looks like a spreadsheet (rows and columns). Lesson 06.\n- **Matplotlib** (and often **Seaborn** built on top of it) — turning numbers into charts you can actually look at. Lesson 07. *(Verify current Seaborn's exact function names against its own docs before relying on them — plotting library APIs do shift between versions.)*\n- **scikit-learn** — ready-made classical machine learning algorithms (Phase 2 of this course).\n- **PyTorch** — the framework for building and training deep learning models, i.e. neural networks (Phase 3). *(TensorFlow/Keras is a real alternative you will see mentioned around Lesson 35 — verify which one a given tutorial or job actually uses, since both remain in active use.)*\n- **Hugging Face Transformers** — a library for loading and running pretrained language models without training one from zero yourself (Phase 4 onward).\n- **LangChain / LlamaIndex** — frameworks for wiring a language model together with your own data and tools (Phase 6 onward). *(Flag this one hardest for verification — these frameworks change their APIs unusually often; always check current docs before trusting a specific function name from any source, including this course.)*\n\nYou are not installing any of these yet — `customer-support-assistant/requirements.txt` stays empty until the lesson that first imports one for real. ## Part 2 — What a notebook actually is\n\nA **notebook** (the `.ipynb` file this lesson lives in) is a document made of **cells** — small blocks that are either **markdown** (formatted text, like this one) or **code**. What makes it different from a normal Python file is what sits behind it: a **kernel**, which is just the name for the one running Python process that keeps every notebook's variables alive in memory as you go.\n\nThat matters because of what it enables: **you run cells one at a time, in any order you choose, and the kernel remembers every variable from every cell you have already run** — even after you move on to write and run a completely different cell below it. Run the next cell, and its variable stays alive in the kernel from that point on."

```python
favorite_number = 7\nprint(favorite_number)
```

The next cell uses `favorite_number`, which the kernel now remembers from the cell above:"

```python
doubled = favorite_number * 2\nprint(doubled)
```

Here is the experiment-and-iterate workflow, made concrete: go back up to the `favorite_number = 7` cell, change the `7` to `21`, and re-run **only that one cell** — not the whole notebook from the top. Then come back down and re-run this `doubled` cell **without changing a single character of its code**. Its output changes from `14` to `42` anyway, because the kernel already forgot the old value of `favorite_number` the moment you re-ran the cell above with the new one.\n\nThis is the entire reason notebooks exist for this kind of work: a plain `.py` script forces you to re-run the *whole file* top to bottom every time you change one line. A notebook lets you change one small piece — one assumption, one number, one line of a data-cleaning step — and re-run just the cells affected by it, seeing the new result in seconds instead of minutes. AI work is mostly this: try something, look at the result, adjust, try again — which is exactly the loop a notebook is built for.\n\nOne habit this course will keep repeating: before you run a cell, predict what it will print. Then run it and check. Being right builds real understanding; being wrong and figuring out why is even more useful."
