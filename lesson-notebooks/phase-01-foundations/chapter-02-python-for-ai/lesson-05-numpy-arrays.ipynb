# Lesson 05 — NumPy and Thinking in Arrays

**What this notebook does:** compares a plain Python list to a NumPy array on the same small set of numbers — ticket text lengths — to show why AI code is written in arrays instead of loops. It builds up from a 1D array, to shape, to a 2D array, to aggregate functions, to filtering with a boolean mask.

## Step 1 — Import NumPy

The convention almost everyone uses is `import numpy as np` — `np` is just a short nickname so you don't have to type `numpy` in front of every single function call. It is not required by Python, only a near-universal habit you will see in every course, tutorial, and real project.

```python
import numpy as np
```

## Step 2 — A plain list versus a NumPy array

Here are the character lengths of four support tickets, first as an ordinary Python list, then as a NumPy array built from that same list with `np.array(...)`. They print almost the same way, but `type(...)` shows they are genuinely different kinds of object underneath.

```python
ticket_lengths = [12, 45, 7, 23]
np_lengths = np.array(ticket_lengths)

print(ticket_lengths)
print(np_lengths)
print(type(ticket_lengths))
print(type(np_lengths))
```

## Step 3 — Doubling every length: a loop versus a vectorized operation

Suppose every length needs to be doubled. With a plain list, Python has no built-in way to do that in one shot, so the usual way is a loop (written below as a list comprehension, which is just a compact loop) — Python revisits each number one at a time, checks its type, and computes `x * 2` for it, one element at a time, all the way through.

With a NumPy array, `np_lengths * 2` does the whole thing at once: the loop still technically happens, but it happens inside NumPy's own fast, compiled code instead of Python's slower interpreter, and you never write the loop yourself. Applying one operation to an entire array in a single step like this is called a **vectorized operation** — "vector" here just means "a list of numbers treated as one unit," so a vectorized operation is one instruction applied to a whole batch of numbers at once, instead of an instruction repeated by hand once per number. This difference feels small with 4 numbers. It is the difference between seconds and hours once the array holds millions of numbers, which is the normal size for real AI data.

```python
doubled_loop = [x * 2 for x in ticket_lengths]
doubled_vectorized = np_lengths * 2

print(doubled_loop)
print(doubled_vectorized)
```

## Step 4 — Shape: how many numbers, arranged how

Every NumPy array has a **shape**: a tuple of numbers describing how many elements it holds along each dimension. `np_lengths` is a flat list of 4 numbers, so its shape is `(4,)` — one dimension, 4 elements along it. That single trailing comma is real Python tuple syntax for "a tuple with just one item," not a typo.

Real data is usually not flat like this — it is a table. Imagine two features per ticket instead of one: its length, and how many `!` characters it contains. Laid out as a table, that is 4 tickets (rows) by 2 features (columns) — a **2D array**, meaning an array with two dimensions: rows and columns, exactly like a spreadsheet. Its shape is `(4, 2)`: 4 rows, 2 columns.

```python
print(np_lengths.shape)

ticket_features = np.array([
    [12, 0],
    [45, 2],
    [7, 0],
    [23, 1],
])
print(ticket_features)
print(ticket_features.shape)
```

## Step 5 — Aggregate functions: turning an array into one number

A **vectorized aggregate function** collapses a whole array down to a single summary number in one call: `.sum()` adds every element together, `.mean()` gives the average (the sum divided by how many elements there are), and `.max()` gives the largest value. Each one is doing, internally, exactly the loop-and-accumulate work you would otherwise write by hand — `.sum()` is really "start a running total at 0, then add every element to it" — but NumPy has already written and optimized that loop for you.

```python
print(np_lengths.sum())
print(np_lengths.mean())
print(np_lengths.max())
```

## Step 6 — A boolean mask: filtering without a loop

Writing `np_lengths > 20` does not raise an error and does not return `True` or `False` — it compares *every element* to 20 at once and returns a brand-new array of `True`/`False` values, one per element. An array of `True`/`False` values like this is called a **boolean mask**: "boolean" is just the technical name for a `True`-or-`False` value, and "mask" describes how it is used next — laid over the original array like a stencil, keeping only the positions marked `True`.

That is exactly what `np_lengths[np_lengths > 20]` does: it uses the mask to pull out only the elements where the mask is `True`, with no `if` statement and no loop written anywhere. This pattern — build a mask with a comparison, then index with that mask — is how filtering is done throughout NumPy and pandas, and it is worth recognizing now before pandas leans on it constantly in Lesson 06.

```python
mask = np_lengths > 20
print(mask)
print(np_lengths[mask])
```
